## 绘制折线图

```python
import matplotlib.pyplot as plt
import numpy as np

# 数据集
x = -np.linspace(-5, 5, 40)
y = -np.sin(x)
z = np.arange(40)

fig = plt.figure()
ax3d = fig.add_subplot(projection="3d")

ax3d.plot(x, y, z)
```

## 绘制面图

```python
import matplotlib.pyplot as plt
import numpy as np

# 生成一维的 array
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

# 根据一维的点生成二维网格的点
x, y = np.meshgrid(x, y)

# 绘制椭圆抛物面
z = x ** 2 / 3 + y ** 2 / 4

fig = plt.figure()
ax3d = fig.add_subplot(projection="3d")
ax3d.plot_surface(x, y, z)
```

## 绘制网格图

```python
import matplotlib.pyplot as plt
import numpy as np

# 生成一维的 array
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

# 根据一维的点生成二维网格的点
x, y = np.meshgrid(x, y)

# 绘制椭圆抛物面
z = x ** 2 / 3 + y ** 2 / 4

fig = plt.figure()
ax3d = fig.add_subplot(projection="3d")
ax3d.plot_wireframe(x, y, z)
```

## 绘制轮廓图

也就是以一定的间隔，用某个垂直于坐标轴的平面截取，将截线绘制出来

```python
import matplotlib.pyplot as plt
import numpy as np

# 生成一维的 array
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
x, y = np.meshgrid(x, y)
z = x ** 2 / 3 + y ** 2 / 4

fig = plt.figure()
ax3d = fig.add_subplot(projection="3d")

# 截线是沿着 y 方向的
ax3d.contour(x, y, z, zdir="y")
```

## 三维向量场

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
z = np.linspace(-5, 5, 10)
x, y, z = np.meshgrid(x, y, z)
u = - x / 10
v = - y / 10
w = - z / 10

fig = plt.figure()
ax3d = fig.add_subplot(projection="3d")
ax3d.quiver(x, y, z, u, v, w)
```

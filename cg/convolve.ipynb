### 计算离散卷积

```python
import numpy as np
import matplotlib.pyplot as plt

def conv(n):
    f = np.repeat(1/8, 8)
    h = f
    for _ in range(n):
        h = np.convolve(h, h)
    return h

fig = plt.figure(figsize=(8, 8))

ax1 = fig.add_subplot(221)
ax1.bar(np.arange(-3, 4 + 1), conv(0))
ax1.set_title(r"$PMF\ of\ S_1$")

ax2 = fig.add_subplot(222)
ax2.bar(np.arange(-3 * 2 ** 1, 4 * 2 ** 1 + 1), conv(1))
ax2.set_title(r"$PMF\ of\ S_2$")

ax3 = fig.add_subplot(223)
ax3.bar(np.arange(-3 * 2 ** 2, 4 * 2 ** 2 + 1), conv(2))
ax3.set_title(r"$PMF\ of\ S_4$")

ax4 = fig.add_subplot(224)
ax4.bar(np.arange(-3 * 2 ** 4, 4 * 2 ** 4 + 1), conv(4))
ax4.set_title(r"$PMF\ of\ S_{16}$")
```

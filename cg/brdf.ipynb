### Normal Distribution

不同粗糙度下的 NDF 性质，使用 GGX

```python
import numpy as np
import matplotlib.pyplot as plt


def ndf(theta, alpha):
    theta = np.deg2rad(theta)
    cos_theta = np.cos(theta)
    alpha2 = alpha ** 2
    return alpha2 / np.pi / np.power(cos_theta ** 2 * (alpha2 - 1) + 1, 2)

# 准备数据
theta = np.linspace(-90, 90, 180)
alpha1 = 0.3
alpha2 = 0.6
alpha3 = 0.9

# 准备画图
fig = plt.figure()
ax = fig.add_subplot(111)

# fixme: 画图
ax.plot(theta, ndf(theta, alpha1), label=r"$\alpha=0.3$")
ax.plot(theta, ndf(theta, alpha2), label=r"$\alpha=0.6$")
ax.plot(theta, ndf(theta, alpha3), label=r"$\alpha=0.9$")

# fixme: 标题
ax.set_title("$NDF$")

# fixme: 横纵坐标
ax.set_xlabel(r"$\theta(n,\ h)$")

# fixme: 显示图例
plt.legend(loc='upper left')
```

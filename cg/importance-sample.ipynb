### 低差异序列
- Van Der Corput 方法进行 radical inversion 运算
- 根据 radical inverse 得到 Halton 序列和 Hammersley 序列

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12, 4))

def gen_VDC(b: int):
    """
    radical inversion van der corput
    b: base
    """
    def VDC(x: int):
        base = 1.0 / b
        res = 0
        while (x > 0):
            value = x % b
            x = x // b
            res += base * value 
            base /= b
        return res 
    return VDC

vVDC_2 = np.vectorize(gen_VDC(2), otypes=[float])   # VDC, base = 2
vVDC_3 = np.vectorize(gen_VDC(3), otypes=[float])   # VDC, base = 3

# 准备数据
N = 100
x = np.arange(1, N + 1)
x_inverse = x / N
y_2 = vVDC_2(x)
y_3 = vVDC_3(x)

# radical inversion Van der Corput
ax1 = fig.add_subplot(131)
ax1.scatter(x, y_2)
ax1.set_title(r"$radical\ inversion\ Van\ der\ Corput$")
ax1.set_xlabel(r"$n$")

# Halton sequence
ax2 = fig.add_subplot(132)
ax2.scatter(y_2, y_3)
ax2.set_title(r"$Halton\ sequence$")
ax2.set_xlabel(r"$Van\ der\ Corput\ base = 2$")
ax2.set_ylabel(r"$Van\ der\ Corput\ base = 3$")

# Hammersley sequence
ax3 = fig.add_subplot(133)
ax3.scatter(x_inverse, y_2)
ax3.set_title(r"$Hammersley\ sequence$")
ax3.set_xlabel(r"$\frac{i}{N}$")
ax3.set_ylabel(r"$Van\ der\ Corput\ base = 2$")
```

### GGX 重要性采样

```python
import numpy as np
import matplotlib.pyplot as plt

rough = 0.4
alpha = rough * rough
y = np.linspace(0, 1, 20)
theta = np.arccos(np.sqrt((1 - y) / (1 + (alpha * alpha - 1) * y)))

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(y, theta/np.pi * 180)
ax.set_xlabel(r"$\xi \sim U(0,\ 1)$")
ax.set_ylabel(r"$\theta \sim\ GGX?$")
```

#### GGX 相关的 pdf 和 CDF

研究一下 GGX 相关的 pdf 和 CDF，看看 GGX 重要性采样的函数是如何得出的

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12, 4))


a = 0.5
y = np.linspace(0, 1, 100)
cos_theta = np.sqrt((1 - y) / (1 + (a * a - 1) * y))
theta = np.arccos(cos_theta) / np.pi * 180.0

# GGX importance sample
ax1 = fig.add_subplot(131)
ax1.plot(y, theta)
ax1.set_xlabel(r"$\xi \sim\ U(0,\ 1)$")
ax1.set_ylabel(r"$\theta \sim\ GGX$")

x = np.linspace(0, np.pi /2.0, 100)
F = (1 - np.cos(x) ** 2) / (np.cos(x) ** 2 * (a * a - 1) + 1)

# CDF
ax2 = fig.add_subplot(132)
ax2.plot(x, F)
ax2.set_xlabel(r"$\theta(radias)$")
ax2.set_ylabel(r"$CDF$")

f_up = 2 * np.sin(x) * np.cos(x) + (a * a - 1) * 2 * np.cos(x) * np.sin(x)
f_dn = (np.cos(x) ** 2 * (a * a - 1) + 1) ** 2
f = f_up / f_dn

# pdf
ax3 = fig.add_subplot(133)
ax3.plot(x, f)
ax3.set_xlabel(r"$\theta(radias)$")
ax3.set_ylabel(r"$pdf$")
```

#### GGX D 乘以 cos(theta)

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12, 4))

a = 0.5
theta = np.linspace(0, 90, 100)
cos_theta = np.cos(np.deg2rad(theta))
D = a ** 2 / np.pi / (((a * a - 1) * cos_theta ** 2 + 1) ** 2)
ax1 = fig.add_subplot(131)
ax1.plot(theta, D)

p = D * cos_theta
ax2 = fig.add_subplot(132)
ax2.plot(theta, p)
```
